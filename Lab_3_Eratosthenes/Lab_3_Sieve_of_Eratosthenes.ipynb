## Algorithm 1

```python
def sieve_of_eratosthenes_1(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        if c[i]:
            j = 2 * i
            while j <= n:
                c[j] = False
                j += i
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

## Algorithm 2

```python
def sieve_of_eratosthenes_2(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        j = 2 * i
        while j <= n:
            c[j] = False
            j += i
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

## Algorithm 3

```python
def sieve_of_eratosthenes_3(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        if c[i]:
            j = i + 1
            while j <= n:
                if j % i == 0:
                    c[j] = False
                j += 1
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

## Algorithm 4

```python
def sieve_of_eratosthenes_4(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        j = 1
        while j < i:
            if i % j == 0:
                c[i] = False
                break
            j += 1
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

## Algorithm 5

```python
def sieve_of_eratosthenes_5(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        j = 2
        while j <= m.sqrt(i):
            if i % j == 0:
                c[i] = False
                break
            j += 1
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```
