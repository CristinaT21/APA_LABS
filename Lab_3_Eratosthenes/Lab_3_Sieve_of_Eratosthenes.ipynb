## Algorithm 1

```python
def sieve_of_eratosthenes_1(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        if c[i]:
            j = 2 * i
            while j <= n:
                c[j] = False
                j += i
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

## Algorithm 2

```python
def sieve_of_eratosthenes_2(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        j = 2 * i
        while j <= n:
            c[j] = False
            j += i
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

## Algorithm 3

```python
def sieve_of_eratosthenes_3(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        if c[i]:
            j = i + 1
            while j <= n:
                if j % i == 0:
                    c[j] = False
                j += 1
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

## Algorithm 4

```python
def sieve_of_eratosthenes_4(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        j = 2
        while j < i:
            if i % j == 0:
                c[i] = False
                break
            j += 1
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

## Algorithm 5

```python
def sieve_of_eratosthenes_5(n):
    c = [True] * (n + 1)
    c[0], c[1] = False, False

    i = 2
    while i <= n:
        j = 2
        while j <= m.sqrt(i):
            if i % j == 0:
                c[i] = False
                break
            j += 1
        i += 1

    primes = []
    for i in range(2, n + 1):
        if c[i]:
            primes.append(i)

    return primes
```

```python
headers=['Sieve of Eratosthenes term number',  'Execution time(wall time)', 'Execution time(CPU time)']
```

```python
list_ex = []
table = []
for i in list_n:
    """
    loop for calculating the execution time
    and including the results in a table
    """
    start = time.time()
    start2 = time.process_time()
    nr = sieve_of_eratosthenes_1(i)
    end = time.time()
    end2 = time.process_time()
    # time in milliseconds
    ex_time = (end - start) * 1000
    ex_time_list.append(ex_time)
    ex_time2 = (end2 - start2) * 1000
    table.append([i, ex_time, ex_time2])

print(f'Algorithm 1')
print(tabulate(table, headers=headers, tablefmt='fancy_grid'))
t = [el[1] for el in table]
list_ex.append(list(t))
plt.plot(list_n, list_ex[0], '-og')
plt.xlabel("array length")
plt.ylabel("execution time (milliseconds)")
plt.title(f"Algorithm 1")
plt.ticklabel_format(useOffset=False, style='plain')
plt.show()
```
