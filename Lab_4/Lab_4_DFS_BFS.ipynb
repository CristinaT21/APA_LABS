```python
# import libraries
import math as m
import time
import matplotlib.pyplot as plt
from tabulate import tabulate
from collections import deque
```

```python
# implement Depth First Search
def dfs(st_node, fin_node, tree):
    visited = set()  # keep track of visited nodes

    def dfs_alg(node, end_node):
        visited.add(node)
        if node == end_node:
            return True

        for neighbor in tree[node]:
            if neighbor not in visited:
                if dfs_alg(neighbor, end_node):
                    return True
        return False

    if dfs_alg(st_node, fin_node):
        return True
    else: return False
```

```python
# implement Breadth First Search
def bfs(st_node, fin_node, tree):
    visited = set()  # keep track of visited nodes
    queue = deque([st_node])  # initialize queue with start node

    while queue:
        node = queue.popleft()
        if node == fin_node:
            return True
        visited.add(node)
        for neighbor in tree[node]:
            if neighbor not in visited:
                queue.append(neighbor)
    return False
```

```python
headers=['Node',  'Execution time(wall time)']
```

```python
list_n = ['D', 'K', 'L', 'T', 'a']
```

```python
range5 = [1, 2, 3, 4, 5]
```

```python
list_ex = []
```

```python
table = []
ex_time_list = []
for i in range(len(list_n)):
    """
    loop for calculating the execution time
    and including the results in a table
    """
    start = time.time()
    time.sleep(0.000000001)
    for node in list_n[:i+1]:
        nr = dfs('A', node, balanced_tree)
    end = time.time()
    # time in milliseconds
    ex_time = (end - start) * 1000
    ex_time_list.append(ex_time)
    table.append([i, ex_time])

print(f'DFS for balanced tree')
print(tabulate(table, headers=headers, tablefmt='fancy_grid'))
t = [el[1] for el in table]
list_ex.append(list(t))
plt.plot(range5, list_ex[0], '-og')
plt.xlabel("number of nodes")
plt.ylabel("execution time (milliseconds)")
plt.title(f"DFS for balanced tree")
plt.ticklabel_format(useOffset=False, style='plain')
plt.show()
```
